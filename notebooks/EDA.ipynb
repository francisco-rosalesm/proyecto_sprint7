```python
#Carga de librerias
import pandas as pd
import plotly.express as px
```

```python
#cargamos el archivo de datos
car_data = pd.read_csv('../vehicles_us.csv') # leer los datos
car_data.name='car_data'

def vista_datos(tabla):
    print("Informacion basica de la tabla: ",tabla.name, "\n")
    tabla.info(show_counts = True)
    print("Filas Duplicadas ",tabla.duplicated().sum(),"\n")
    print("Nulos: ",tabla.isna().sum(),"\n")
    print(tabla.describe())
    display(tabla.head())
    
#llamamos la funcion
vista_datos(car_data)
```

Observamos que tiene datos vacios en algunas columnas: model_year, cylinders, odometer, paint_color y  en is_4wd. Tambine que algunas columnas tiene un tipo de dato incorrecto que debemos cambiar. Price de int a float, model_year de float a object, de cylinders de floar a int, is_4wd de float a boolean, y date_posted de object a date.

```python
#primero trabajaremos con la primera columna Price, cambio de int a float
car_data['price']=car_data['price'].astype(float)

#ahora vamos por model_year, primero de float a int, rellenamos primero los nan
car_data['model_year'].fillna(0, inplace=True)
car_data['model_year']=car_data['model_year'].astype(int)

#ahora vamos por cylinders
car_data['cylinders'].fillna(0, inplace=True)
car_data['cylinders']=car_data['cylinders'].astype(int)

#ahora es turno de paint_color, solo rellenaremos los nan.
car_data['paint_color'].fillna("unknow", inplace=True)

#para is_4wd vamos a rellenar primero con 0, y despues cambiamos a bollean
car_data['is_4wd'].fillna(0, inplace=True)
car_data['is_4wd']=car_data['is_4wd'].astype(bool)

#vamos a cambiar date_posted de object a date
car_data['date_posted'] = pd.to_datetime(car_data['date_posted'])

#verificamos con la funcion vista_datos
vista_datos(car_data)
```

```python
#crear un histograma con los datos de vehicles 

fig = px.histogram(car_data, x="model_year", 
                   title="Autos posteados por año de modelo") # crear un histograma
fig.show() # crear gráfico de dispersión

#crear un grafico de dispersion 

fig = px.scatter(car_data, x="odometer", y="price", title="Precio de vehiculos por kilometraje") # crear un gráfico de dispersión
fig.show() # crear gráfico de dispersión
```
